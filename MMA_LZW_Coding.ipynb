```python
def lzw_encode(data):
    dictionary = {chr(i): i for i in range(256)}
    result = []
    current_code = 256
    current_sequence = ""

    for symbol in data:
        current_sequence += symbol
        if current_sequence not in dictionary:
            result.append(dictionary[current_sequence[:-1]])
            dictionary[current_sequence] = current_code
            current_code += 1
            current_sequence = symbol

    if current_sequence in dictionary:
        result.append(dictionary[current_sequence])

    return result
```

```python
def lzw_decode(encoded_data):
    dictionary = {i: chr(i) for i in range(256)}
    result = []
    current_code = 256
    previous_code = encoded_data[0]
    previous_entry = dictionary[previous_code]
    result.append(previous_entry)

    for code in encoded_data[1:]:
        if code in dictionary:
            entry = dictionary[code]
        elif code == current_code:
            entry = previous_entry + previous_entry[0]
        else:
            raise ValueError("Invalid LZW encoded data")

        result.append(entry)
        dictionary[current_code] = previous_entry + entry[0]
        current_code += 1
        previous_entry = entry

    return ''.join(result)
```

```python
# Example
if __name__ == "__main__":
    original_message = "encoding and decoding  in python for  LZW coding"

    # LZW Encoding
    encoded_message = lzw_encode(original_message)
    print("Encoded message:", encoded_message)

    # LZW Decoding
    decoded_message = lzw_decode(encoded_message)
    print("Decoded message:", decoded_message)
```
